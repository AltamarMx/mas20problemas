```python
coef = np.polyfit(x, y, deg=5)
p = np.poly1d(coef)

plt.plot(x,y)
plt.plot(x, p(x))
```

```python
def ajuste(grado):
    coef = np.polyfit(x,y,deg=grado)
    p = np.poly1d(coef)
    plt.plot(x,y)
    plt.plot(x, p(x))
    plt.ylim(-200,1000)
    #plt.title(p)
```

```python
ajuste(4)
```

```python
interact(ajuste, grado=(0,10))
```
