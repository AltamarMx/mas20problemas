## [Calplot](https://github.com/tomkwok/calplot)

Calplot creates heatmaps from Pandas time series data.

Plot Pandas time series data sampled by day in a heatmap per calendar year using matplotlib.



```python
import pandas as pd
import calplot 
```

```python
f = '../data/001_raw/temixco.csv'
esolmet = pd.read_csv(f,index_col=0,parse_dates=True)
esolmet
```

```python
f = '../data/001_raw/RUOA.txt'
ruoa = pd.read_csv(f,index_col=0,parse_dates=True,skiprows=[0,2],sep='\t',dayfirst=True)
ruoa.rename(columns={'Temp_Avg':'To'},inplace=True)
ruoa.head()
```

## Energ'ia solar diaria visualizada

```python
calplot.calplot((esolmet.Ig)*(1/6),cmap='jet')
```

```python
calplot.calplot(esolmet.To,how='max',cmap='jet',vmax=37 )
calplot.calplot(ruoa.To,how='max',cmap='jet',vmax=37 )
```
