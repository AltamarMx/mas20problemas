## [Calplot](https://github.com/tomkwok/calplot)

```python
import pandas as pd
import calplot 
```

```python
f = '../data/001_raw/temixco.csv'
esolmet = pd.read_csv(f,index_col=0,parse_dates=True)
esolmet
```

```python
f = '../data/001_raw/RUOA.txt'
ruoa = pd.read_csv(f,index_col=0,parse_dates=True,skiprows=[0,2],sep='\t',dayfirst=True)
ruoa.rename(columns={'Temp_Avg':'To'},inplace=True)
ruoa.resample('600S').interpolate(method='time')
```

```python
ruoa
```

```python
ruoa.To.plot()
```

```python
dif = ruoa.To - esolmet.To
dif.resample('D').mean()
```

```python
dif
```

```python
calplot.calplot(dif)
```
