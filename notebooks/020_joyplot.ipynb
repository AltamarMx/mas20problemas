```python
import joypy
import seaborn as sns
import pandas as pd
from matplotlib import cm
```

```python
df = pd.read_csv('../data/001_raw/temixco.csv',index_col=0,parse_dates=True)
```

```python
df['Year']  = df.index.year
df['Month'] = df.index.month
df['Day']   = df.index.day_of_year
```

```python
df
```

```python
fig, axes = joypy.joyplot(df, by="Month", column="To",  range_style='own', 
                          grid="y", linewidth=1, legend=False, figsize=(12,5),colormap=cm.autumn_r)
```
