```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

import matplotlib.ticker as ticker
```

```python
nombres = ['tiempo','Tsi','Tso','q']
ti = pd.read_csv('../data/001_raw/time-dependent.csv',index_col=0,names=nombres)
ti.index = ti.index/60
```

```python
nombres = ['tiempo','Tsi','Tso','q']
td = pd.read_csv('../data/001_raw/time-dependent.csv',index_col=0,names=nombres)
td.index = td.index/60
```

```python
#Options
params = {'text.latex.preamble': '\\usepackage{lmodern}',
          'text.usetex' : True,
          'font.size' : 10,
          'font.family' : 'lmodern'
          }
plt.rcParams.update(params) 
```

```python
fig = plt.figure(constrained_layout=True,figsize=(4.8,4))

gs = GridSpec(3, 4, figure=fig)

ax1 = fig.add_subplot(gs[0:1, 0:2])
ax2 = fig.add_subplot(gs[0:1, 2:4])
ax3 = fig.add_subplot(gs[1:3, :])




ax1.set_ylabel('Temperature $[^\circ C]$')
ax1_2 = ax1.twinx() 
ax1_2.set_ylabel('Heat flux [$W/m^2$]')
ax1.set_xlabel('Time [min]')
ax1.set_ylim(10,60)
ax1.set_xlim(-0.5,25)
ax1_2.set_ylim(0,80)
ax1.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax1.yaxis.set_major_locator(ticker.MultipleLocator(10))
ax1.text(2,50,'a)')
# ax1.grid()
ax1.plot(  ti.Tso,'r--',label='$T_{so}$')
ax1.plot(  ti.Tsi,'b--',label='$T_{si}$')
ax1_2.plot(ti[::60].q,'k--',label='$q_o$')
    
    
    
ax2.set_ylabel('Temperature $[^\circ C]$')
ax2_2 = ax2.twinx()  # instantiate a second axes that shares the same x-axis
ax2_2.set_ylabel('Heat flux [$W/m^2$]')
ax2.set_xlabel('Time [min]')
ax2.set_ylim(25,75)
ax2.set_xlim(-0.5,25)
ax2_2.set_ylim(20,100)
ax2.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax2.yaxis.set_major_locator(ticker.MultipleLocator(10))
ax2.text(2,65,'b)')
# ax2.grid()    
ax2.plot(  td.Tso,'r--',label='$T_{so}$')
ax2.plot(  td.Tsi,'b--',label='$T_{si}$')
ax2_2.plot(td.q,'k-',label='$q_o$')
    
    
    

ax3.set_ylabel('Heat flux [$W/m^2$]')
ax3.set_xlabel('Time [min]')
ax3.set_ylim(20,100)
ax3.set_xlim(-0.5,25)
ax3.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax3.yaxis.set_major_locator(ticker.MultipleLocator(20))
ax3.text(0.8,87,'c)')
# # ax2.grid()    
ax3.plot(td.q,'k-',label='$q_o$')
ax3.plot(td.q*2-100,'k--',label='$q_i$')
  
    
    
    
# plt.savefig('../figs/analysis.pdf',dpi=1000)
plt.savefig('../figs/analysis.png',dpi=1000)
# tikzplotlib.save("../figs/analysis.pgf")
```
