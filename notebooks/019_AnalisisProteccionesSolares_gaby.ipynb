```python
con =Este_con.iloc[:,0:4].loc['2006-05'].resample('D').max().resample('M').mean()
sin =Este_sin.iloc[:,0:4].loc['2006-05'].resample('D').max().resample('M').mean()
```

```python
con['N1LAB'] = Oeste_con.iloc[:,0:1].loc['2006-05'].resample('D').max().resample('M').mean()
sin['N1LAB'] = Oeste_sin.iloc[:,0:1].loc['2006-05'].resample('D').max().resample('M').mean()
con
```

```python
N1 = pd.concat([con,sin])
N1
```

#### Graficar

```python
N1.T.plot.bar()
```

```python
N1.T.plot.bar(color=['b','grey'],figsize=(8,4) )

plt.title("N1 con y sin protecciones solares")
plt.ylabel('Tmax Mayo [°C]',size = 12)
plt.legend(['con','sin'])
plt.grid(axis='y', linestyle='--')
plt.ylim(24, 36)

```
