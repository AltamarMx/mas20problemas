```python
import pandas as pd
import numpy as np
```

```python
nombres = np.genfromtxt('../data/001_raw/nombres_DATA.txt',
                        dtype='U',delimiter=',')
# nombres
data = pd.read_csv('../data/001_raw/DATA.csv',names=nombres,
                  index_col=0,parse_dates=True)
data = data.resample('10Min').interpolate(method='time')
```

```python
data.Ti.plot()
```

```python
ruoa = pd.read_csv('../data/001_raw/RUOA.txt',skiprows=[0,2],
                  delimiter='\t',index_col=0,parse_dates=True)
ruoa.plot(figsize=(12,3))
```

```python
ruoa = pd.read_csv('../data/001_raw/RUOA.txt',skiprows=[0,2],
                  delimiter='\t',index_col=0,parse_dates=True,dayfirst=True)
ruoa = ruoa.resample('600S').interpolate(method='time')
ruoa.plot(figsize=(12,3))
```

```python
datos = pd.concat([ruoa,data],axis=1)
# datos.Ti.plot()
datos.columns
```

```python
datos.Ti.plot
```

```python
datos.to_csv('../data/002_processed/experimento.csv')
```
