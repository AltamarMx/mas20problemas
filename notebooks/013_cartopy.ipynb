```python
lats = np.loadtxt("../data/001_raw/lats_10.csv")
lons = np.loadtxt("../data/001_raw/lons_10.csv")

u10 = np.loadtxt("../data/001_raw/era_u10.csv")
v10 = np.loadtxt("../data/001_raw/era_v10.csv")

vels = np.sqrt(u10**2+v10**2)
```

https://scitools.org.uk/cartopy/docs/v0.15/matplotlib/intro.html

```python
plt.figure(figsize=(8,16))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-100,-80,33,17])
ax.coastlines(resolution="10m")

#ax.add_feature(cfeature.LAND, facecolor="antiquewhite", zorder=1)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.STATES.with_scale("10m"))
plot=ax.pcolormesh(lons,lats,vels, cmap="jet",transform=ccrs.PlateCarree())
plt.colorbar(plot, shrink=0.4)
```

```python
plt.figure(figsize=(8,16))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-94,-86,16,23])
ax.coastlines(resolution="10m")

#ax.add_feature(cfeature.LAND, facecolor="antiquewhite", zorder=1)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.STATES.with_scale("10m"))
plot=ax.pcolormesh(lons,lats,vels, cmap="jet",transform=ccrs.PlateCarree())
ax.quiver(lons, lats, u10, v10, scale=250)
plt.colorbar(plot, shrink=0.4)
```
