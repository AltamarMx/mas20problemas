```python
import seaborn as sns
import pandas as pd
from matplotlib import cm
```

```python
df = pd.read_csv('../data/001_raw/temixco.csv',index_col=0,parse_dates=True)
```

```python
df['Year']  = df.index.year
df['Month'] = df.index.month
df['Day']   = df.index.day_name()
```

```python
df
```

```python
df["tiempo"] = df.index.hour + df.index.minute/60

sns.lineplot(x='tiempo',data=df,y="To",ci="sd",err_style="band",palette="red")
```

```python
df["tiempo"] = df.index.hour + df.index.minute/60

sns.lineplot(x='tiempo',data=df,y="To",
             ci="sd",
             err_style="band",
            hue="Day")
```

```python
df["tiempo"] = df.index.hour + df.index.minute/60

sns.lineplot(x='tiempo',data=df,y="To",
             ci="sd",
             err_style="band",
            hue="Month")
```

```python
sns.relplot(
    data=df,
    x="tiempo", y="To",
    col="Day",
    kind="line")
```
