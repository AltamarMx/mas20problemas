# Realiza el mapa de calor de la ecuación de sensación térmica

$$ IST = -8.7846+1.6113T+2.3385HR-0.1461THR-0.0123T^2-0.0164HR^2+0.0022T^2HR+0.0007THR^2-0.000003T^2HR^2 $$

Donde STC es el Índice de Sensación Térmica por calor, T es la temperatura del aire ambiente en grados Celsius y HR es la humedad relativa del aire en %. En la AEMET este cálculo se lleva a cabo para valores de la temperatura ambiente superiores a 26 ºC y humedades superiores al 40%

https://www.tiempo.com/ram/289402/la-sensacion-termica-y-sus-calculos-por-la-aemet/#:~:text=En%20AEMET%2C%20los%20valores%20de,de%20la%20superficie%20del%20suelo

```python
def ist(T,HR):
    return (-8.7846+1.6113*T+2.3385*HR-0.1461*T*HR-0.0123*(T**2)-0.0164*(HR**2)+0.0022*(T**2)*HR+0.0007*T*(HR**2)-0.000003*(T**2)*(HR**2))
```

```python
ist(30,78)
```

```python
humedad = np.arange(40,101,5)
humedad
```

```python
len(humedad)
```

```python
temperatura = np.linspace(26,44,13)
len(temperatura)
```

```python
temperatura
```

```python
ist(temperatura, humedad)
```

```python
ist(26, 40)
```

```python
np.meshgrid(temperatura, humedad)
```

```python
x_mesh, y_mesh = np.meshgrid(temperatura, humedad)
```

```python
ist(x_mesh, y_mesh)
```

```python
sensacion = ist(x_mesh, y_mesh)
```

```python
sensacion
```

```python
type(sensacion)
```

```python
sensacion[sensacion<=50]
```

```python
fig, ax = plt.subplots(figsize=(8,8))
p = ax.imshow(sensacion, cmap="hot_r")
ax.set_xticks(np.arange(13))
ax.set_xticklabels(temperatura)
ax.set_yticks(np.arange(13))
ax.set_yticklabels(humedad)
ax.set_xlabel("Temperatura [°C]")
ax.set_ylabel("Humedad relativa [%]")
fig.colorbar(p, shrink=0.8, label="Sensación térmica [°C]")
```
